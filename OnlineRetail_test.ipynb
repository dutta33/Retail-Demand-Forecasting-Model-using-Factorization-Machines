```python
def DataTransform(filename):
    data = []
    y = []
    users=set()
    items=set()
#     print(range(len(filename)))
    for line in range(len(filename)):
#         print(df2_UK.loc[line,"CustomerID"])
        data.append({"UserID": str(filename.loc[line,"CustomerID"]), "StockCode": str(filename.loc[line,"StockCode"])}) 
        y.append(float(filename.loc[line,"Quantity"]))
        users.add(filename.loc[line,"CustomerID"])
        items.add(filename.loc[line,"StockCode"])

    return (data, np.array(y), users, items)
```
